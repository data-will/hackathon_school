**Code Machine learning avec un KNN proche voisin qui permet de récuperer les 3 sports les plus proches des caractéristiques sportives de l'individu :**

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors

def sport_predict2(user_performance):
    # Initialisation du X en incluant seulement les variables numériques
    X = df_sport.select_dtypes(include='number')
    
    # Initialisation du modèle KNN
    distanceKNN = NearestNeighbors(n_neighbors=3).fit(X)

    # Transformation de la performance de l'utilisateur en une ligne de données
    user_df = pd.DataFrame([user_performance], columns=X.columns)

    # Recherche des sports les plus proches
    distance, indices = distanceKNN.kneighbors(user_df)

    # Affichage des sports les plus proches à l'utilisateur
    print("Les sports les plus proches de votre performance sont :")
    for idx in indices:
        print(df_sport.loc[idx]['SPORT'])

# Demande à l'utilisateur d'entrer sa performance
user_skill1 = float(input("Entre 0 et 10, Entrez votre niveau d'Endurance: "))
user_skill2 = float(input("Entre 0 et 10, Entrez votre niveau de Force : "))
user_skill3 = float(input("Entre 0 et 10, Entrez votre niveau de Puissance : "))
user_skill4 = float(input("Entre 0 et 10, Entrez votre niveau de Vitesse : "))
user_skill5 = float(input("Entre 0 et 10, Entrez votre niveau d'Agilité  : "))
user_skill6 = float(input("Entre 0 et 10, Entrez votre niveau de Souplesse : "))
user_skill7 = float(input("Entre 0 et 10, Entrez votre niveau de Sang-froid : "))
user_skill8 = float(input("Entre 0 et 10, Entrez votre niveau de Résistance: "))
user_skill9 = float(input("Entre 0 et 10, Entrez votre niveau de Coordination main-œil : "))
user_skill10 = float(input("Entre 0 et 10, Entrez votre niveau de Stratégie : "))



# Stocke les performances de l'utilisateur dans une liste
user_performance = [user_skill1, user_skill2, user_skill3, user_skill4, user_skill5,
                    user_skill6, user_skill7, user_skill8, user_skill9, user_skill10]


sport_predict2(user_performance)
```

**Code streamlit avec Machine learning KNN proche voisin afin de trouver les 3 plus proches sport en fonction des caractéristiques sportives de l'individu :**

```python
import streamlit as st
import pandas as pd
from sklearn.neighbors import NearestNeighbors


# Charger les données
df_sport = pd.read_csv('C:\\Users\\33660\\Desktop\\hakaton\\sport_sans_catégorie.csv')

# Fonction pour prédire les sports les plus proches
def sport_predict2(user_performance):
    # Initialisation du X avec les variables numériques
    X = df_sport.select_dtypes(include='number').drop('Unnamed: 0', axis=1)
    
    # Initialisation du modèle KNN
    distanceKNN = NearestNeighbors(n_neighbors=3).fit(X)

    # Transformation de la performance de l'utilisateur en une ligne de données
    user_df = pd.DataFrame([user_performance], columns=X.columns)

    # Recherche des sports les plus proches
    distance, indices = distanceKNN.kneighbors(user_df)

    # Affichage des sports les plus proches à l'utilisateur
    st.write("Les sports les plus proches de votre performance sont :")
    for idx in indices:
        st.write(df_sport.loc[idx]['SPORT'])

# code streamlit pour l'interraction utilisateur :
st.title("Prédiction de sports")
st.write("Veuillez entrer vos niveaux de performance :")

user_skill1 = st.slider("Endurance", 0, 10, 1)
user_skill2 = st.slider("Force", 0, 10, 1)
user_skill3 = st.slider("Puissance", 0, 10, 1)
user_skill4 = st.slider("Vitesse", 0, 10, 1)
user_skill5 = st.slider("Agilité", 0, 10, 1)
user_skill6 = st.slider("Souplesse", 0, 10, 1)
user_skill7 = st.slider("Sang-froid", 0, 10, 1)
user_skill8 = st.slider("Résistance", 0, 10, 1)
user_skill9 = st.slider("Coordination main-œil", 0, 10, 1)
user_skill10 = st.slider("Stratégie", 0, 10, 1)

# liste contenant chaque caractéristiques choisie par l'utilisateur
if st.button("Prédire"):
    user_performance = [user_skill1, user_skill2, user_skill3, user_skill4, user_skill5,
                        user_skill6, user_skill7, user_skill8, user_skill9, user_skill10]
    
    sport_predict2(user_performance)
```
